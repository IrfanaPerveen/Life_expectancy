```python
# creat new vari(life_expectancy_cy_df) having information with respect to country.
life_expectancy_cy_df = life_expectancy_df.groupby('Country')[life_expectancy_df.columns].mean()
life_expectancy_cy_df.drop(columns=['Year'], inplace=True)
life_expectancy_cy_df.reset_index(inplace=True)
life_expectancy_cy_df
```

```python
# taking status with respect to country.
country_status_df = life_expectancy_df.groupby('Country')['Status'].min()
country_status_df = country_status_df.reset_index()
country_status_df
```

```python
# merging the above two tables on common column country.
life_expectancy_cy_df = life_expectancy_cy_df.merge(country_status_df, on='Country')
life_expectancy_cy_df
```

```python
# Filter for developed and developing countries
df_developed = life_expectancy_df[life_expectancy_df['Status'] == 'Developed']
df_developing = life_expectancy_df[life_expectancy_df['Status'] == 'Developing']

# Group by year and calculate average life expectancy for developed and developing countries
avg_life_expectancy_developed = df_developed.groupby('Year')['Life expectancy '].mean()
avg_life_expectancy_developing = df_developing.groupby('Year')['Life expectancy '].mean()

plt.figure(figsize=(10, 6)) # Set figure size
plt.plot(avg_life_expectancy_developed.index, avg_life_expectancy_developed, label='Developed') # Plot line for developed countries
plt.plot(avg_life_expectancy_developing.index, avg_life_expectancy_developing, label='Developing') # Plot line for developing countries
plt.xlabel('Year') 
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy by Year (Developed vs Developing)') 
plt.legend()
plt.grid(True) 
plt.show()
```

```python

import scipy.stats as stats
# Load the dataset
life_expectancy_df =pd.read_csv('Life_Expectancy_Data.csv')

# Filter for developed and developing countries
df_developed = life_expectancy_df[life_expectancy_df['Status'] == 'Developed']
df_developing = life_expectancy_df[life_expectancy_df['Status'] == 'Developing']

# Perform independent samples t-test
t_stat, p_value = stats.ttest_ind(df_developed['Life expectancy '].dropna(), df_developing['Life expectancy '].dropna(), equal_var=False)
#t_stat, p_value = stats.ttest_ind(df_developed['Schooling'].dropna(), df_developing['Schooling'].dropna(), equal_var=False)
# Print t-statistic and p-value
print('t-statistic:', t_stat)
print('p-value:', p_value)

# Check for significance at a chosen alpha level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print('There is a significant difference in the mean of developed and developing countries.')
else:
    print('There is no significant difference in the mean of developed and developing countries.')
```

```python
# Perform independent samples t-test
t_stat, p_value = stats.ttest_ind(df_developed['Schooling'].dropna(), df_developing['Schooling'].dropna(), equal_var=False)

# Print t-statistic and p-value
print('t-statistic:', t_stat)
print('p-value:', p_value)

# Check for significance at a chosen alpha level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print('There is a significant difference in the mean of developed and developing countries.')
else:
    print('There is no significant difference in the mean of developed and developing countries.')
```

```python
# Perform independent samples t-test
t_stat, p_value = stats.ttest_ind(df_developed['Income composition of resources'].dropna(), df_developing['Income composition of resources'].dropna(), equal_var=False)

# Print t-statistic and p-value
print('t-statistic:', t_stat)
print('p-value:', p_value)

# Check for significance at a chosen alpha level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print('There is a significant difference in the mean of developed and developing countries.')
else:
    print('There is no significant difference in the mean of developed and developing countries.')
```

```python

# Drop rows with missing values in 'Schooling' and 'Life expectancy' columns
df_developed_clean = df_developed.dropna(subset=['Schooling', 'Life expectancy '])
df_developing_clean = df_developing.dropna(subset=['Schooling', 'Life expectancy '])

# Calculate Pearson correlation coefficient for 'Schooling' in developed countries
corr_coef_developed, p_value_developed = stats.pearsonr(df_developed_clean['Schooling'], df_developed_clean['Life expectancy '])
print('Pearson correlation coefficient for Schooling in developed countries:', corr_coef_developed)
print('p-value:', p_value_developed)

# Calculate Pearson correlation coefficient for 'Schooling' in developing countries
corr_coef_developing, p_value_developing = stats.pearsonr(df_developing_clean['Schooling'], df_developing_clean['Life expectancy '])
print('Pearson correlation coefficient for Schooling in developing countries:', corr_coef_developing)
print('p-value:', p_value_developing)
```
